## 📥 Paso 1: Importar las funciones

## 🧪 Paso 2: Probar el algoritmo

Ejecuta las siguientes celdas para probar cada función:

### Prueba 1: Ordenar una lista

```python
# Crear una lista de prueba
mi_lista = [5, 2, 9, 1, 7]

print(f"Lista original: {mi_lista}")
print(f"Lista ordenada: {ordenar_lista(mi_lista)}")
```

### Prueba 2: Buscar un elemento

```python
mi_lista = [5, 2, 9, 1, 7]
elemento_a_buscar = 9

posicion = buscar_elemento(mi_lista, elemento_a_buscar)

if posicion != -1:
    print(f"✅ Elemento {elemento_a_buscar} encontrado en posición {posicion}")
else:
    print(f"❌ Elemento {elemento_a_buscar} no encontrado")
```

### Prueba 3: Invertir una lista

```python
mi_lista = [5, 2, 9, 1, 7]

print(f"Lista original: {mi_lista}")
print(f"Lista invertida: {invertir_lista(mi_lista)}")
```

### Prueba 4: Calcular suma y promedio

```python
mi_lista = [5, 2, 9, 1, 7]

print(f"Lista: {mi_lista}")
print(f"Suma: {suma_lista(mi_lista)}")
print(f"Promedio: {promedio_lista(mi_lista):.2f}")
```
